```python
import pandas as pd

df = pd.read_csv("cl_flipkart_data.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

-----heatmap show storng correlation between variables.....

```python
sns.heatmap(
    df[['selling_price', 'ratting_number', 'reviews']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x=df['selling_price'], showfliers=False)
plt.title("Price Distribution (Outlier Hidden)")
plt.xlabel("selling price")
plt.show()
```

```python
sns.barplot(x='Price_Band', y='ratting_number', data=df)
plt.title("Average Rating by Price Band")
plt.show()
```

```python
sns.scatterplot(x='selling_price', y='ratting_number', data=df)
plt.title("Price vs Rating")
plt.show()
```

```python
#----Budget phones dominate listings
#----Higher price does NOT always mean higher rating
#----Reviews positively correlate with popularity
#----Premium phones have fewer but higher-quality ratings
```
